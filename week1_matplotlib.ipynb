```python
import matplotlib.pyplot as plt

x = [5, 4, 3, 2, 1]
y = [10, 20, 55, 35, 85]

plt.plot(x, y)
plt.title("My First Matplotlib Graph")
plt.xlabel("X Axis")
plt.ylabel("Y Axis")
plt.show()
```
