```python
import pandas as pd
df= pd.read_csv("/content/session4_dummy_students_data.csv.xls")
print(df)
```

```python
#convert math score column to numpy array

import numpy as np
math_score= df["math"].to_numpy()
print(math_score)
print(type(math_score))
print(math_score.mean())
```

```python
#for science score column
science_score= df["science"].to_numpy
print(science_score)
```

```python
#calculate average marks across all subjects
science_score = df["science"].to_numpy()
math_score = df["math"].to_numpy()
english_score = df["english"].to_numpy()

avg = df[["science", "math", "english"]].mean(axis=1)


top_names = df[avg > 85]["name"]

low_attendance_names = df[df["attendance_percent"] < 75]["name"]

mathura_names = df[df["city"] == "Mathura"]["name"]
print(top_names)
print(low_attendance_names)
print(mathura_names)
```

```python
df["average"] = (df["math"] + df["science"] + df["english"]) / 3

df["result"] = "Fail"
df.loc[df["average"] >= 40, "result"] = "Pass"

print("Class Average:", df["average"].mean())

print("Highest Average:", df["average"].max())
print("Lowest Average:", df["average"].min())

top_students = df[df["average"] == df["average"].max()]
print("Top Student:")
print(top_students["name"])

print("Attendance Below 75:")
print(df[df["attendance_percent"] < 75]["name"])

print("Assignments Below 6:")
print(df[df["assignments_submitted"] < 6]["name"])
```
